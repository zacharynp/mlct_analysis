```python
import cv2
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from autoliv_util.autolivutil import ImageProcessing, ImageAnalysis, SqlManagement, VideoProcessing
sns.set()
```

```python
path = r"temp/images_repositioned"
video_folders = [os.path.join(path, x) for x in os.listdir(path)]
video_folders
```

```python
df = pd.DataFrame()
# for video_folder in video_folders:
# video_folder = video_folders[0]
for video_folder in video_folders:
    images, data = ImageAnalysis.track_cushion_analysis(images_folder=video_folder)
    this_df = pd.DataFrame(data, columns=['x', 'y', 'w', 'h', 'area'])
    this_df['video'] = os.path.split(video_folder)[1]
    this_df['Time'] = np.arange(0, 40.25, 0.25)
    df = pd.concat([df, this_df])
df.head()
```

```python
df = df[['area', 'Time', 'video']]
```

```python
pivot = df.pivot(index='Time', columns=['video'], values=['area'])
```

```python
df[df['Time'] < 4.0]
```

```python
plt.figure(figsize=(15, 10))
new_df = df.set_index('Time')
for video in new_df['video'].unique():
    this_df = new_df[new_df['video'] == video]
    this_df['area'] = df['area'].apply(lambda x: 0 if x > 20000 else x)
    sns.lineplot(data=this_df, x=this_df.index, y='area')
plt.xlabel('Time (ms)')
plt.ylabel('Area')
plt.show()
```
