```python
import seaborn as sns
import pandas as pd
import cv2
import matplotlib.pyplot as plt
frames = [17, 17.5, 18, 19, 20]
sns.set()
```

```python
df = pd.read_csv('Scores_17.0ms.csv')
```

```python
df.head()
```

```python
sns.histplot(df['Score'], bins=10)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
for i in range(10):
    row = df.iloc[i]
    image_path = row['File Path']
    score = row['Score']
    image = cv2.imread(image_path)
    plt.subplot(4, 4, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(image)
    plt.xlabel('{0:5.2f}'.format(score / 100000))
plt.show()
```
